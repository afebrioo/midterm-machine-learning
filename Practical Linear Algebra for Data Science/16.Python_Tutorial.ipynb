## 1. Dasar Pemrograman Python

Python adalah bahasa pemrograman yang mudah dibaca dan sangat populer dalam data science. Berikut adalah beberapa elemen dasarnya:

```python
# Variabel dan Tipe Data Dasar
a = 5
b = 3.14
nama = "Aljabar Linier"
print(f"a: {a}, tipe: {type(a)}")
print(f"b: {b}, tipe: {type(b)}")
print(f"nama: {nama}, tipe: {type(nama)}")
```

```python
# List di Python
my_list = [10, 20, 30, 40]
print("List asli:", my_list)
print("Elemen pertama:", my_list[0])
print("Slicing (2 elemen pertama):", my_list[:2])
```

```python
# Loop dan Conditional (Percabangan)
for x in my_list:
    if x > 25:
        print(f"{x} lebih besar dari 25")
    else:
        print(f"{x} kurang dari atau sama dengan 25")
```

```python
# Mendefinisikan Fungsi
def pangkat_dua(x):
    return x ** 2

print("Hasil pangkat dua dari 4:", pangkat_dua(4))
```

## 2. NumPy untuk Aljabar Linier

NumPy (`Numerical Python`) adalah library standar untuk komputasi array dan matriks di Python.

```python
import numpy as np

# Membuat Vektor (Array 1D)
v = np.array([1, 2, 3])
print("Vektor v:", v)
print("Ukuran/Shape v:", v.shape)
```

```python
# Membuat Matriks (Array 2D)
A = np.array([[1, 2], [3, 4]])
print("Matriks A:\n", A)
print("Shape A:", A.shape)
```

```python
# Transpose Matriks
print("A Transpose (A.T):\n", A.T)
```

```python
# Perkalian Titik (Dot Product) Vektor dan Perkalian Matriks
w = np.array([4, 5, 6])
dot_product = np.dot(v, w)
print(f"Dot product v dan w: {dot_product}")

B = np.array([[5, 6], [7, 8]])
C = np.matmul(A, B)
print("Hasil perkalian matriks A * B:\n", C)
```

```python
# Matriks Khusus: Zeros, Ones, Identity (Eye)
print("Matriks Nol (3x3):\n", np.zeros((3, 3)))
print("Matriks Identitas (3x3):\n", np.eye(3))
```

## 3. Matplotlib untuk Visualisasi

Matplotlib adalah library utama untuk visualisasi data dan grafik di Python.

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Membuat Plot Garis Sederhana
x = np.linspace(-5, 5, 100)
y = x ** 2

plt.figure(figsize=(6, 4))
plt.plot(x, y, 'r-', linewidth=2, label='$y = x^2$')
plt.grid(True, linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Kurva Kuadratis')
plt.legend()
plt.show()
```

```python
# Menggambar Vektor dengan plt.arrow
plt.figure(figsize=(5, 5))
plt.arrow(0, 0, 3, 4, head_width=0.3, head_length=0.3, color='blue', length_includes_head=True)
plt.plot(0, 0, 'ko') # Titik pangkal di (0,0)
plt.grid(True, linestyle='--')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal')
plt.title('Visualisasi Vektor [3, 4]')
plt.show()
```

```python
# Menampilkan Matriks Acak Sebagai Gambar (Citra Grayscale)
matrix_images = np.random.randn(10, 10)
plt.figure(figsize=(4, 4))
plt.imshow(matrix_images, cmap='gray')
plt.title('Visualisasi Matriks Acak')
plt.colorbar()
plt.show()
```
