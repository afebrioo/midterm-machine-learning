## 1. Apa itu Aljabar Linier dan Mengapa Mempelajarinya?

Aljabar linier adalah cabang matematika yang berurusan dengan **vektor** (vector), **matriks** (matrix), dan **transformasi linier** (linear transformation). Sejarah aljabar linier dimulai sejak abad ke-17 di dunia Barat (dan jauh sebelumnya di Tiongkok) sebagai alat untuk menyimpan koordinat geometris (oleh René Descartes) dan menyelesaikan sistem persamaan linier (oleh Carl Friedrich Gauss).

Di era modern, aljabar linier menjadi pilar utama komputasi. Komputer dirancang untuk melakukan kalkulasi pada matriks dan vektor secara sangat cepat dan efisien. Hampir semua algoritma machine learning modern (mulai dari regresi linier, PCA, hingga Deep Learning) dibangun di atas fondasi aljabar linier.

## 2. Aljabar Linier Tradisional vs. Aljabar Linier Modern

Pendekatan pembelajaran aljabar linier dalam buku ini berfokus pada **Aljabar Linier Modern (Komputasional)**:

| Karakteristik | Aljabar Linier Tradisional | Aljabar Linier Modern (Komputasional) |
| :--- | :--- | :--- |
| **Fokus Utama** | Teorema, pembuktian matematis, ruang vektor dimensi tak hingga. | Kode program, data riil, visualisasi, dan aplikasi praktis. |
| **Alat Utama** | Kertas, pensil, dan manipulasi aljabar manual. | Komputer, Python, NumPy, dan visualisasi grafik. |
| **Tujuan** | Pemahaman matematis murni akademik. | Menyelesaikan masalah komputasi, kecerdasan buatan (AI), grafik, dan data science. |

Buku ini tidak menuntut Anda menghafal pembuktian rumus matematis yang rumit, melainkan membangun intuisi geometri dan komputasi melalui kode Python.

## 3. Aplikasi Aljabar Linier dalam Data Science

Berikut adalah beberapa contoh nyata aplikasi aljabar linier dalam data science:
1. **Representasi Data Citra/Gambar:** Gambar digital direpresentasikan sebagai matriks angka piksel (2D untuk grayscale, 3D untuk warna RGB).
2. **Reduksi Dimensi (PCA):** Menggunakan dekomposisi nilai singular (SVD) dan dekomposisi eigen untuk mengurangi fitur data yang kompleks.
3. **Sistem Rekomendasi (Collaborative Filtering):** Menggunakan faktorisasi matriks untuk memprediksi preferensi pengguna.
4. **Deep Learning (Neural Networks):** Pembelajaran jaringan saraf tiruan didominasi oleh perkalian matriks bobot (*weights matrix multiplication*).
