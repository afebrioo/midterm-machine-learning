### 1. NumPy
NumPy adalah salah satu pustaka paling mendasar untuk komputasi ilmiah dalam Python. ### 2. SciPy
SciPy adalah kumpulan fungsi matematika untuk komputasi ilmiah dalam Python. SciPy menyediakan pustaka aljabar linier tingkat lanjut, optimasi fungsi matematika, pemrosesan sinyal, fungsi matematika khusus, dan distribusi statistik.

Modul yang paling penting dari SciPy bagi kita adalah `scipy.sparse` yang menyediakan **Matriks Jarang (Sparse Matrices)**. Matriks jarang digunakan jika dataset kita berupa tabel 2D yang sebagian besar elemennya bernilai nol (misalnya representasi bag-of-words pada pemrosesan teks), sehingga menghemat memori penyimpanan secara signifikan.

```python
from scipy import sparse

# Membuat array NumPy 2D dengan diagonal bernilai 1, dan elemen lain bernilai 0 (Matriks Identitas)
eye = np.eye(4)
print("NumPy array:\n", eye)
```

```python
# Mengonversi array NumPy tersebut ke matriks sparse SciPy dalam format CSR (Compressed Sparse Row)
# Hanya elemen yang tidak bernilai nol yang disimpan di memori
sparse_matrix = sparse.csr_matrix(eye)
print("\nSciPy sparse CSR matrix:\n", sparse_matrix)
```

```python
# Cara alternatif membuat matriks sparse menggunakan representasi COO (Coordinate)
data = np.ones(4)
row_indices = np.arange(4)
col_indices = np.arange(4)
eye_coo = sparse.coo_matrix((data, (row_indices, col_indices)))
print("COO representation:\n", eye_coo)
```

## Aplikasi Pertama: Klasifikasi Spesies Iris

### Mengenal Data (Meet the Data)

Kasus pertama kita adalah klasifikasi klasikal: dataset bunga **Iris**. Dataset ini memiliki pengukuran fisik sepal (kelopak luar) dan petal (mahkota dalam) dari 150 sampel bunga Iris. Ini merupakan contoh dari **Supervised Learning** (karena kita melatih model dengan data berlabel spesies yang sudah diketahui) dan **Multiclass Classification** (karena target memiliki 3 kelas berbeda).

### Mengukur Keberhasilan: Data Pelatihan dan Pengujian (Training and Testing Data)

Kita ingin membangun model machine learning yang dapat memprediksi spesies Iris untuk data baru yang belum terlabel. Namun, sebelum model tersebut digunakan, kita harus mengetahui tingkat keandalannya. Apakah kita bisa mempercayai prediksinya?

Kita tidak boleh mengevaluasi model menggunakan dataset yang sama dengan data pelatihan. Mengapa? Karena model dapat dengan mudah menghafal seluruh set data pelatihan dan akan selalu memprediksi label yang benar (mengakibatkan *overfitting*), tanpa tahu apakah model dapat digeneralisasikan pada data baru di luar sana.

Untuk mengatasinya, kita membagi 150 sampel data kita menjadi dua bagian:
1. **Training Data (Training Set):** Digunakan untuk membangun dan melatih model (75% dari total data).
2. **Test Data (Test Set / Hold-out Set):** Digunakan untuk mengukur akurasi dan kinerja model yang dilatih (25% dari total data).

scikit-learn menyediakan fungsi `train_test_split` untuk mengacak dataset dan membaginya untuk kita secara acak. Pengacakan dikendalikan oleh parameter `random_state` untuk memastikan hasil pembagian yang konsisten saat dijalankan ulang.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
```

Dalam konvensi matematika, fitur masukan dilambangkan dengan huruf besar `X` (karena data berupa matriks 2D) dan label keluaran dilambangkan dengan huruf kecil `y` (karena data berupa vektor 1D) yang diturunkan dari rumusan fungsi $f(X) = y$.

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

### Membangun Model Pertama: k-Nearest Neighbors (k-NN)

Sekarang, kita membangun model klasifikasi sesungguhnya menggunakan algoritma **k-Nearest Neighbors (k-NN)**. Cara kerja algoritma ini adalah:
1. Saat menerima data pengujian baru, model mencari $k$ titik data pelatihan terdekat di ruang dimensi fitur.
2. Model memprediksi kelas data baru tersebut berdasarkan suara mayoritas kelas dari tetangga terdekat tersebut.

Dalam implementasi ini, kita memilih jumlah tetangga terdekat ($k$) bernilai 1. Di dalam scikit-learn, algoritma k-NN diimplementasikan melalui kelas `KNeighborsClassifier` di modul `neighbors`.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

Untuk melatih model k-NN dengan data pelatihan, kita memanggil metode `.fit()` yang menerima argumen fitur pelatihan `X_train` dan target label pelatihan `y_train`.

```python
knn.fit(X_train, y_train)
```

### Membuat Prediksi (Making Predictions)

Setelah melatih model, kita dapat membuat prediksi untuk bunga Iris baru yang belum terlabel. Bayangkan kita menemukan bunga Iris di alam liar dengan:
- Panjang sepal = 5 cm
- Lebar sepal = 2.9 cm
- Panjang petal = 1 cm
- Lebar petal = 0.2 cm

Kita masukkan data tersebut ke dalam array NumPy 2D karena scikit-learn mengharuskan input berupa matriks 2D.

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape:", X_new.shape)
```

```python
prediction = knn.predict(X_new)
print("Prediction:", prediction)
print("Predicted target name:",
      iris_dataset['target_names'][prediction])
```

Model memprediksi bahwa bunga baru tersebut merupakan spesies **Setosa** (indeks kelas 0).

### Evaluasi Model (Evaluating the Model)

Untuk mengukur kinerja model secara objektif, kita menggunakan data pengujian (Test Set: `X_test` dan `y_test`). Data tes ini tidak pernah dilihat oleh model selama proses pelatihan. Kita dapat mengukur persentase akurasi model dengan membandingkan prediksi model terhadap label asli.

```python
y_pred = knn.predict(X_test)
print("Test set predictions:\n", y_pred)
```

```python
# Menggunakan NumPy mean untuk menghitung tingkat akurasi rata-rata
print("Test set score (manual mean): {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
# Menggunakan metode score bawaan dari objek estimator knn
print("Test set score (estimator score): {:.2f}".format(knn.score(X_test, y_test)))
```

Akurasi model kita pada set data pengujian adalah **0.97** (atau 97%), menunjukkan model kita dapat dipercaya untuk memprediksi data Iris baru dengan tingkat kegagalan yang sangat rendah.

### Ringkasan Alur Kerja Pemodelan Klasifikasi

Berikut ini adalah gabungan baris kode esensial dari alur kerja pemodelan klasifikasi yang telah kita lakukan di atas secara ringkas:

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```
